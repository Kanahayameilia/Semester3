```python
from collections import Counter
import re

# Baca dataset (contoh file CSV atau TXT)
filename = "spam.csv"
with open(filename, 'r', encoding='latin-1') as f:
    text = f.read()

# Preprocessing teks
text = text.lower()  # huruf kecil semua
text = re.sub(r'[^a-z\s]', '', text)  # hapus tanda baca

# Split kata
words = text.split()

# Map step: buat pasangan (kata, 1)
mapped = [(word, 1) for word in words]

# Reduce step: hitung jumlah setiap kata
counter = Counter()
for word, count in mapped:
    counter[word] += count

# Ambil 20 kata terbanyak
most_common_words = counter.most_common(20)
print("20 kata paling sering muncul:")
for word, count in most_common_words:
    print(f"{word}: {count}")
```

```python
import pandas as pd
from collections import Counter
import re

# Baca CSV
df = pd.read_csv("spam.csv", encoding='latin-1')  # ganti encoding sesuai dataset
# Kolom teks biasanya 'v2'
texts = df['v2'].astype(str)

all_words = []
for text in texts:
    # Preprocessing: huruf kecil, hapus tanda baca
    clean_text = re.sub(r'[^a-z\s]', '', text.lower())
    all_words.extend(clean_text.split())

word_count = Counter(all_words)
most_common_words = word_count.most_common(10)
print("10 kata paling sering muncul:")
for word, count in most_common_words:
    print(f"{word}: {count}")
```
